# Seaborn Plots

## Distribution Plots

1. DistPlot
2. JointPlot
3. PairPlot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in Feature Section

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### JoinPlot

A join plot allows to study the relationship between 2 numeric variables.The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

### PairPlot

A "pair plot" also known as scatterplot, in which one variable in the same data rows matched with another variable's.

```python
sns.pairplot(df,hue='sex')
```

### DistPlot

Dist plot helps us to check the distribution of the columns features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],bins=10,kde=False)
```

## Categorical Plots

1. Countplot
2. BarPlot
3. BoxPlot
4. ViolinPlot

```python
df.head()
```

### CountPlot

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='day',data=df)
```

### BarPlot 

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

### BoxPlot

A box plot is a graph that represents information from a five-number summary.

```python
sns.boxplot('day','total_bill',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

### ViolinPlot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='sex',split=True)
```
